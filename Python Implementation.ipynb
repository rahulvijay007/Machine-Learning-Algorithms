# Preperation for modelling

```python
df.drop(['residual sugar', 'free sulfur dioxide', 'pH'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
sns.countplot(df['quality'])
```

```python
bins = (2, 6.5, 8)
group_names = ['bad', 'good']
df['quality'] = pd.cut(df['quality'], bins = bins, labels = group_names)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
new_quality = LabelEncoder()
```

```python
df['quality'] = new_quality.fit_transform(df['quality'])
```

```python
df.head()
```

```python
df['quality'].value_counts()
```

```python
x = df.drop(['quality'], axis = 1)
y = df['quality']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

# Scalling

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

# Modelling

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
svc_model.fit(x_train, y_train)
```

```python
y_pred = svc_model.predict(x_test)
```

```python
cv_scores = cross_val_score(estimator = svc_model, X = x_train, y = y_train, cv = 10)
cv_scores.mean()
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(cnf_matrix, annot = True, cmap = 'Spectral', fmt = '6')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print(classification_report(y_test, y_pred))
```
